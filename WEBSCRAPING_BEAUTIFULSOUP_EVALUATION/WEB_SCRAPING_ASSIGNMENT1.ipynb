<h2>QUESTION 1 ANSWER ⬇️ </h2>

```python
url1 = 'https://www.imdb.com/list/ls056092300/'
```

```python
def imdb(url1):
    page1 = requests.get(url1)
    
    soup = BeautifulSoup(page1.content, 'html.parser')
    
    listofmovies = []
    for match in soup.find_all('h3', class_='lister-item-header'):
        for match1 in match.find_all('a'):
            listofmovies.append(match1.text)
    
    listofratings = []
    for match in soup.find_all('div', class_='ipl-rating-widget'):
        for match1 in match.find('span', class_ = 'ipl-rating-star__rating'):
            listofratings.append(match1.text)
    
    listofyear = []
    for match in soup.find_all('span', class_ = 'lister-item-year text-muted unbold'):
        listofyear.append(match.text)
    
    dfmovies = pd.DataFrame(columns = ['Name', 'Rating', 'Year of release'])
    
    dfmovies['Name'] = listofmovies
    
    dfmovies['Rating'] = listofratings
    
    dfmovies['Year of release'] = listofyear
    
    return dfmovies
    
```

```python
imdb(url1)
```

<h2>QUESTION 6 ANSWER ⬇️ </h2>

```python
url6 = 'https://www.bewakoof.com/bestseller?sort=popular'
```

```python
def func6(url6):

    page6 = requests.get(url6)
    
    page6
    
    soup6 = BeautifulSoup(page6.content, 'html.parser')
    
    productname = []
    for match in soup6.find_all('h2'):
        print(match.text)
        productname.append(match.text)
    
    imageurls = []
    for match in soup6.find_all('div',class_='productCardImg'):
        print(match.img['src'])
        imageurls.append(match.img['src'])
    
    discountedprices = []
    for match in soup6.find_all('div', class_ = 'discountedPriceText'):
        print(match.text)
        discountedprices.append(match.text)
    
    df6 = pd.DataFrame(columns = ['Product Name','Image URL', 'Price'])
    
    df6['Product Name'] = productname
    
    df6['Image URL'] = imageurls
    
    df6['Price'] = discountedprices
    
    return df6
```

<h1>QUESTION 7 ANSWER ⬇️</h1>

```python
url7 = 'https://www.cnbc.com/world/?region=world'
```

```python
def func7(url7):

    page7 = requests.get(url7)
    
    soup7 = BeautifulSoup(page7.content, 'html.parser')
    
    headings = []
    for match in soup7.find_all('div', class_ = 'RiverHeadline-headline RiverHeadline-hasThumbnail'):
        print(match.text)
        headings.append(match.text)
    
    newslinks = []
    for match in soup7.find_all('div', class_ = 'RiverHeadline-headline RiverHeadline-hasThumbnail'):
        print(match.a['href'])
        newslinks.append(match.a['href'])
    
    dates = []
    for match in soup7.find_all('time'):
        print(match.text)
        dates.append(match.text)
    
    df7 = pd.DataFrame(columns=['Heading','Article Link','Date'])
    
    df7['Heading'] = headings
    
    df7['Article Link'] = newslinks
    
    df7['Date'] = dates[:20]
    
    return df7
```

```python
func7(url7)
```

<h1>QUESTION 8 ANSWER ⬇️</h1>

```python
url8 = 'https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/'
```

```python
def func8(url8):
    
    page8 = requests.get(url8)
    
    soup8 = BeautifulSoup(page8.content, 'html.parser')
    
    papertitles = []
    for match in soup8.find_all('h2', class_='h5 article-title'):
        papertitles.append(match.text.strip())
    
    articledate = []
    for match in soup8.find_all('p', class_='article-date'):
        articledate.append(match.text)
    
    authors = []
    for match in soup8.find_all('p', class_='article-authors'):
        authors.append(match.text.strip())
    
    df8 = pd.DataFrame(columns=['Paper Title', 'Publish Date', 'Authors'])
    
    df8['Paper Title'] = papertitles
    
    df8['Publish Date'] = articledate
    
    df8['Authors'] = authors
    
    return df8
```

```python
func8(url8)
```
