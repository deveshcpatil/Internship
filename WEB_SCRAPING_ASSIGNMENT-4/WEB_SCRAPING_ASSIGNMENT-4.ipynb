```python
driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a').click()
```

```python
repositorytitle = []
repositorydescription = []
contributorscount = []
languageused = []
```

```python
for match in driver.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]'):
    print(match.text)
    repositorytitle.append(match.text)
```

```python
repositorylinks = []
for match in driver.find_elements(By.XPATH, '//h2[@class="h3 lh-condensed"]/a'):
    print(match.get_attribute('href'))
    repositorylinks.append(match.get_attribute('href'))
```

```python
driver.quit()
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
n = 1
for i in repositorylinks:
    driver.get(i)
    print(f'Now extracting repository No- {n}/{len(repositorylinks)}')
    time.sleep(3)
    try:
        var1 = driver.find_element(By.XPATH, '//div[@class="hide-sm hide-md"]/p').text
        print(var1)
        repositorydescription.append(var1)
    except NoSuchElementException:
        print('------')
        repositorydescription.append('-----')
    try:
        var2 = driver.find_element(By.XPATH, '//div[@class="BorderGrid-row"]/div/ul[@class="list-style-none"]').text
        print(var2)
        languageused.append(var2)
    except NoSuchElementException:
        print('------------')
        languageused.append('-----')
    try:
        var3 = driver.find_element(By.XPATH, '//div[@class="BorderGrid-row"]/div/h2[@class="h4 mb-3"]/a/span[@class="Counter ml-1"]').text
        print(var3)
        contributorscount.append(var3)
    except NoSuchElementException:
        print('****************')
        contributorscount.append('-----')
    print('///////////////////////////////////////////// REPOSITORY OVER')
    print('\n\n')
    n = n + 1
driver.quit()
```

```python
df4 = pd.DataFrame({'Repository title':repositorytitle,
                   'Repository description':repositorydescription,
                    'Contributors count':contributorscount,
                    'Languages used':languageused
                   })
```

```python
df4
```

```python
df4.to_csv('Github Trending Repositories Info.csv', index=False)
```

<h1>QUESTION 8 ANSWER ⬇️ </h1>

```python
driver = webdriver.Chrome()
```

```python
driver.maximize_window()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
driver.find_element(By.XPATH, '//a[@class="btn-primary btn"]').click()
```

```python
from selenium.common.exceptions import ElementClickInterceptedException
```

```python
datasetslinks = []
n=1
while True:
    print('Now extracting links from page no - ', n)
    time.sleep(3)
    for match in driver.find_elements(By.XPATH, '//h2[@class="truncate text-primary"]/a'):
        print(match.get_attribute('href'))
        datasetslinks.append(match.get_attribute('href'))
    try:
        driver.find_element(By.XPATH, '//button[@aria-label="Next Page"]').click()
    except ElementClickInterceptedException:
        print('EXTRACTING LINKS COMPLETE #######################')
        break
    n = n + 1
```

```python
datasetslinks
```

```python
len(datasetslinks)
```

```python
driver.quit()
```

```python
datasetname = []
datatype = []
task = []
attributetype = []
noofinstances = []
noofattributes = []
datasetyear = []
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
n=1
for i in datasetslinks:
    driver.get(i)
    print(f'Now extracting information from dataset {n}')
    time.sleep(1)
    var1 = driver.find_element(By.TAG_NAME, 'h1').text
    print(var1)
    datasetname.append(var1)
    var2 = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p').text
    print(var2)
    datatype.append(var2)
    var3 = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p').text
    print(var3)
    task.append(var3)

    var4 = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p').text
    print(var4)
    attributetype.append(var4)

    var5 = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p').text
    print(var5)
    noofinstances.append(var5)

    var6 = driver.find_element(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p').text
    print(var6)
    noofattributes.append(var6)
    try:
        var7 = driver.find_element(By.TAG_NAME, 'h2').text[-4:]
        print(var7)
        datasetyear.append(var7)
    except NoSuchElementException:
        datasetyear.append('-')
    print('  ')
    n+=1
```

```python
df8 = pd.DataFrame({'Dataset name':datasetname,
                   'Data Type':datatype,
                   'Task':task,
                   'Attribute type':attributetype,
                   'No of instances':noofinstances,
                   'No of attributes':noofattributes,'Year':datasetyear})
```

```python
df8
```

```python
df8.to_csv('UCIDATASETS.csv', index=False)
```
