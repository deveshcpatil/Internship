<h1>QUESTION 2 ANSWER ⬇️ </h1>

```python
brandname = []
productname = []
price = []
expecteddelivery = []
producturl = []
availability = []
exchange = []
```

```python
for i in range(0,3):

    for match in driver.find_elements(By.XPATH, '//div[@data-cy="title-recipe"]/h2/a/span'):
        print(match.text.split(' ')[0])
        brandname.append(match.text.split(' ')[0])
    
    for match in driver.find_elements(By.XPATH, '//div[@data-cy="title-recipe"]/h2/a/span'):
        print(match.text)
        productname.append(match.text)
    
    for match in driver.find_elements(By.XPATH, '//div[@data-cy="price-recipe"]'):
        print(match.text)
        if len(match.text)==0:
            price.append('N/A')
            availability.append('Not Available')
        else:
            price.append(match.text)
            availability.append('Available')
    
    for match in driver.find_elements(By.XPATH, '//div[@data-cy="delivery-recipe"]'):
        if len(match.text)==0:
            expecteddelivery.append('Currently unavailable')
        else:
            print(match.text)
            expecteddelivery.append(match.text)
    
    for match in driver.find_elements(By.XPATH, '//div[@data-cy="title-recipe"]/h2/a'):
        print(match.get_attribute('href'))
        producturl.append(match.get_attribute('href'))

    driver.find_element(By.CLASS_NAME, 's-pagination-next').click()
    time.sleep(3)
```

```python
len(expecteddelivery)
```

```python
expecteddelivery.append('-')
```

```python
exchange = ['-' for i in range(0,70)]
```

```python
df1 = pd.DataFrame({'Brand Name':brandname,
                   'Name of the Product':productname,
                    'Price':price,
                    'Return/Exchange':exchange,
                    'Expected Delivery':expecteddelivery,
                    'Availability':availability,
                    'Product URL':producturl
                   })
```

```python
df1['Price']=df1['Price'].apply(lambda x:x[:7])
```
